```python
a={7,8,9,11}
b={8,12,13}
print(a.union(b),a|b)  #combines all the elememnts of a and b without repeating the common elements
print(a.intersection(b),a&b) #only the common elements
print(a.difference(b),a-b) #all elements of 'a' excluding the common elements
print(b.difference(a),b-a)  #all elements of 'b' excluding the common elements    
print(a.symmetric_difference(b),((a|b)-(a&b))) # combing 2 sets excludinng the common elements 
```

```python
#built-in functions

a={1,2,3,90,'kl'} #adds as one element in the set but not ordered
a.add(99) #adds one value at a time
a
```

```python
a={1,2,3,90,'kl'}
a.update('ghi') #updates the set with the given string as three different elements
a
```

```python
a={1,2,3,90,'kl'} #removes random values ex. lucky draw
b=a.pop() #removes and returns the elememnt cannot predict which element is popped
b
```

```python
a={1,2,3,90,'kl'}
a.clear() #clears the set and returns with empty set
a
```

```python
a={13,12,14,14}
a.remove(14) # removes the given element from the set
a
```

```python
a={12,13,14,12} #discard the given element from the set
a.discard(12)
a
```
