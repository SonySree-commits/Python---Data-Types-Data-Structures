TUPLE  ;Heterogeneous
       ordered
        immutable,
        nested
        duplication allowed

```python
a=() #empty tuple
type(a)
```

```python
a=90,12,#packing
a
```

```python
a,b,c=11,12,13 #unpacking : system first packs the 3 values into one and then unpacks into 3 diff variables
print(a,b,c)
```

```python
a,b,c='909'
print(a,b)
```

```python
a,b,c=909 #cannot unpack because no of variables and values should be equal
a
```

```python
a,b,c,d=1,'sony',1+7j,False #heterogeneous : can assign different data types
print(a,b,c,d)
```

```python
a=(1,2.3,4+6j,True,[12,3,4]) #can also have list inside a tuple 
a 
```

```python
a,b,c = 1,2,3 #ordered supports indexing and slicing
print(a,b,c)
```

```python
a,b,c= 1,1,2 #duplication is allowed
print(a,b,c)
```

```python
a=( 1,2,3) #immutable
a[0] = 34  # does not support item assignment 
a
```

```python
a = (1,2,(2,3))  #nested
a
```

```python
a = 12,12.3,4,56
a
```

```python
#built-in functions

t=(1,2,3,4,1) #gives the count of the value present in the tuple : here 2 1's are present
t.count(1)
```

```python
t=(1,2,3,4,1) #gives the index number of the value
t.index(2)
```

```python
t=(1,2,3,4,1)
len(t) # gives the length of the tuple
```

```python
t=(1,2,3,4,1.5)
min(t) # gives minimum value of the tuple
```

```python
t=(1,2,3,4,81) # gives the maximum value in the tuple
max(t)
```

```python
t=(1,2,3,4.9,1+8j) # sum of the values
sum(t)
```

```python
t=(1,2,3,4,1) # deletes the tuple variables and values
del t
```
