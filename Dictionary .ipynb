             duplication is allowed in values not keys,
             keys need to be unique
             mutable doesn't allow unhashable data types at KEYS , VALUES allow all data types
             Keys are immutable
             nested

```python
a={'c1' : ['pen','paper','sharpner'], 'c2' : ['pen','paper'], 'c3' : ['pen','paper','eraser'], 'c4' : ['sharpner']}
print(a) 
print(a.keys()) #prints the keys here dict_type is the data typr for keys
print(a.values()) #prints the values and dict_values is the data type for values
print(a.items()) # prints the items in tuple 
```

BUILT-IN FUNCTIONS

```python
a={12:13,3:4,5:6}
b={1:111,2:222,3:333}
print(a.get(12)) # gives the value of the key
a.update({12:999})
print (a) #updates the key value
a.update({77:88 })# added at the end
print (a)
print(a.fromkeys(b))# keys of b is taken
print(a.fromkeys(a,b)) # keys of a is taken and whole b is given as value for each key
print(dict.fromkeys(a,b))  #same as above other way to write
print(list(zip(a,b))) # gives the keys in list format
print(dict(zip(a,b))) # gives the keys in dict 
print(set(zip(a,b))) # gives the keys as set
```
