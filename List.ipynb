```python

a=[1,2,3,4,5]
b=a.copy()
print(a,b) #copy means copying the list to another variable and if we change one value of a then it will not get reflected in b
a[3]=999
print(a,b)
```

```python
import copy
a=[1,2,3,4,5]  #same as copy
b=copy.deepcopy(a)
print(a,b)
a[3]=999
print(a,b)
```

```python
a=[1,2,3,[12,13]]
b=a.copy()  # if we change one value of a then it will get reflected in b if the input has nested list
print(a,b)
a[3][1]=999
print(a,b)
```

```python
import copy
a=[1,2,3,[12,13]]
b=copy.deepcopy(a)  #deepcopy doesn't work as copy here changed value in a doesn't reflect in b in nested list also
print(a,b)
a[3][1]=999
print(a,b)
```

copy and deepcopy works the same in normal list but in nested list the chnaged value in a doesn't reflect in b for deep copy
